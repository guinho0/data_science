leitura da planilha correta

```python
import pandas as pd

# Verificar as abas disponíveis
xls = pd.ExcelFile("CTG.xls")
print("Abas disponíveis:", xls.sheet_names)

# Ler a aba correta
df = pd.read_excel(xls, sheet_name="Raw Data")
df.columns = df.columns.str.strip()  # Remove espaços extras
```

remover colunas problemáticas

```python
# 1. Remover colunas que contêm texto ou não são úteis para o modelo
colunas_remover = ['FileName', 'Date', 'SegFile', 'CLASS']
df = df.drop(columns=colunas_remover, errors='ignore')  # ignora caso alguma não exista

# 2. Remover todas as linhas que tenham qualquer valor ausente (NaN)
df = df.dropna()

# 3. Garantir que só restem colunas numéricas (float, int)
df = df.select_dtypes(include=['number'])

# 4. Separar variáveis independentes (X) e a dependente (y)
X = df.drop(columns=['NSP'])  # atributos
y = df['NSP']                 # rótulo

# 5. Verificar se está tudo certo
print("Formato X:", X.shape)
print("Formato y:", y.shape)
print("Valores únicos em y:", y.unique())
```

divisao entre treino e teste

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

função de avaliação dos modelos

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

def avaliar_modelo(modelo, X_test, y_test, nome):
    y_pred = modelo.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    acuracia = accuracy_score(y_test, y_pred)
    precisao = precision_score(y_test, y_pred, average='weighted')
    revocacao = recall_score(y_test, y_pred, average='weighted')

    print(f"\n🔍 {nome}")
    print("Matriz de Confusão:\n", cm)
    print(f"Acurácia:  {acuracia:.4f}")
    print(f"Precisão:  {precisao:.4f}")
    print(f"Revocação: {revocacao:.4f}")
```

KNN com k = 3,5,7,9

```python
from sklearn.neighbors import KNeighborsClassifier

for k in [3, 5, 7, 9]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    avaliar_modelo(knn, X_test, y_test, f"KNN (k={k})")
```

SVM com Kernel linear e RBF

```python
from sklearn.svm import SVC

svm_linear = SVC(kernel='linear')
svm_linear.fit(X_train, y_train)
avaliar_modelo(svm_linear, X_test, y_test, "SVM (Linear)")

svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_train, y_train)
avaliar_modelo(svm_rbf, X_test, y_test, "SVM (RBF)")
```

arvore de decisão com gini e entropia

```python
from sklearn.tree import DecisionTreeClassifier

tree_gini = DecisionTreeClassifier(criterion='gini', random_state=42)
tree_gini.fit(X_train, y_train)
avaliar_modelo(tree_gini, X_test, y_test, "Árvore de Decisão (Gini)")

tree_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)
tree_entropy.fit(X_train, y_train)
avaliar_modelo(tree_entropy, X_test, y_test, "Árvore de Decisão (Entropia)")
```
